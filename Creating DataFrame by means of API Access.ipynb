```python
import json
import requests
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_pesawat_final = pd.DataFrame([])
dateofdeparture = 20200507
for i in range(0,8):
    apps_id = 'xxxxxxx'
    apps_key = 'xxxxxxxxx'
    source = 'SIN'
    destination = 'JKT'
    url_pesawat = 'https://developer.goibibo.com/api/search/?app_id=' + apps_id + '&app_key=' + apps_key + '&format=json&source=' + source + '&destination='+ destination + '&dateofdeparture=' + str(dateofdeparture) + '&seatingclass=E&adults=1&children=0&infants=0&counter=100'  
    data_pesawat = requests.get(url_pesawat)
    dateofdeparture += 1
    for i in range(0,len(data_pesawat.json()['data']['onwardflights'])):
        a = [data_pesawat.json()['data']['onwardflights'][i]]
        b = [data_pesawat.json()['data']['onwardflights'][i]['fare']]
        df_pesawat1 = pd.DataFrame(a)[['origin','destination','airline','depdate']]
        df_pesawat2 = pd.DataFrame(b)['totalfare']
        df_pesawat = pd.concat([df_pesawat1,df_pesawat2],axis='columns')
        df_pesawat_final = df_pesawat_final.append(df_pesawat)
```

```python
df_pesawat_final
```

```python
df_pesawat_final['depdate'] = df_pesawat_final['depdate'].map(lambda x: str(x)[:-5])
df_pesawat_final = df_pesawat_final.reset_index(drop=True)
df_pesawat_final
```

```python
df_pesawat_mean = df_pesawat_final.groupby('depdate',as_index=False).mean().rename(
    columns={'totalfare':'Rata-rata harga (INR)'})
df_pesawat_min = df_pesawat_final.groupby('depdate',as_index=False).min().rename(
    columns={'totalfare':'Harga Terkecil (INR)'})
df_pesawat_max = df_pesawat_final.groupby('depdate',as_index=False).max().rename(
    columns={'totalfare':'Harga Terbesar (INR)'})
df_pesawat = pd.merge(pd.merge(df_pesawat_min,df_pesawat_max,on='depdate'),df_pesawat_mean,on='depdate')
df_pesawat = df_pesawat.rename(columns = {'airline_x': 'Airline Harga Terkecil','airline_y': 'Airline Harga Terbesar'})
df_pesawat = df_pesawat[['depdate','Airline Harga Terkecil','Harga Terkecil (INR)','Airline Harga Terbesar','Harga Terbesar (INR)','Rata-rata harga (INR)']]
```

```python
df_pesawat
```

```python
url_uang = 'https://api.exchangeratesapi.io/latest?base=INR'
data_uang = requests.get(url_uang)
inr_to_idr = data_uang.json()['rates']['IDR']
df_pesawat['Harga Terkecil (IDR)'] = df_pesawat['Harga Terkecil (INR)'].apply(lambda x: x * inr_to_idr)
df_pesawat['Harga Terbesar (IDR)'] = df_pesawat['Harga Terbesar (INR)'].apply(lambda x: x * inr_to_idr)
df_pesawat['Rata-rata harga (IDR)'] = df_pesawat['Rata-rata harga (INR)'].apply(lambda x: x * inr_to_idr)
df_pesawat = df_pesawat[['depdate','Airline Harga Terkecil','Harga Terkecil (INR)','Harga Terkecil (IDR)','Airline Harga Terbesar','Harga Terbesar (INR)','Harga Terbesar (IDR)','Rata-rata harga (INR)','Rata-rata harga (IDR)']]
df_pesawat
```

```python
plt.figure(figsize=(5,5))
plt.plot(df_pesawat['depdate'],df_pesawat['Harga Terkecil (INR)'])
plt.plot(df_pesawat['depdate'],df_pesawat['Harga Terbesar (INR)'])
plt.plot(df_pesawat['depdate'],df_pesawat['Rata-rata harga (INR)'])
plt.xticks(rotation=45)

plt.show()
# seaborn_grid = sns.lmplot('value', 'wage', col='variable', hue='education', data=df_melt, sharex=False)
```
